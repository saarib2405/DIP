```python
import cv2
import matplotlib.pyplot as plt
import textwrap
# load the image and convert to grayscale
inpimage = cv2.imread("/content/lena color.png")
gray = cv2.cvtColor(inpimage, cv2.COLOR_BGR2GRAY)
#Apply Laplacian Edge Detection
laplacian1 = cv2.Laplacian(gray,cv2.CV_64F)
#Apply Inverse Thresholding Laplacian Edge Detection
ret,thresh1 = cv2.threshold(laplacian1,20,255, cv2.THRESH_BINARY_INV)
#Denoising image before edge detection
denoiseimg = cv2.GaussianBlur(gray,(3,3),0)
#Apply Laplacian Edge Detection
laplacian2 = cv2.Laplacian(denoiseimg,cv2.CV_64F)
```

```python
#Apply Inverse Thresholding Laplacian Edge Detection
ret,thresh2 = cv2.threshold(laplacian2,20,255, cv2.THRESH_BINARY_INV)
title1="Grayscale Image"
plt.subplot(331).set_title("\n".join(textwrap.wrap(title1, 20))),plt.imshow(gray,cmap = 'gray'),
plt.axis('off')
title2="Laplacian Edge detection"
plt.subplot(332).set_title("\n".join(textwrap.wrap(title2, 20))),plt.imshow(laplacian1,cmap =
'gray'), plt.axis('off')
title3 = "After Inverse Thresholding"
plt.subplot(333).set_title("\n".join(textwrap.wrap(title3, 20))),plt.imshow(thresh1,cmap =
'gray'), plt.axis('off')
title4 = "Denoised Image"
plt.subplot(337).set_title("\n".join(textwrap.wrap(title4, 20))),plt.imshow(denoiseimg,cmap =
'gray'), plt.axis('off')
title5 = "Laplacian Edge detection on Denoised Image"
plt.subplot(338).set_title("\n".join(textwrap.wrap(title5, 20))),plt.imshow(laplacian2,cmap =
'gray'), plt.axis('off')
title6 = "After Inverse Thresholding"
plt.subplot(339).set_title("\n".join(textwrap.wrap(title6, 20))),plt.imshow(thresh2,cmap =
'gray'), plt.axis('off')
plt.savefig("/content/lena color.png", format="png",dpi=1200)
plt.show()
```
