```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Reading an image in default mode
image = cv2.imread("/content/lena color.png")
#Converting to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#Converting grayscale to binary
ret,thresh = cv2.threshold(gray,70,255, cv2.THRESH_BINARY_INV)
# Creating kernel
kernel = np.ones((5, 5), np.uint8)
print(kernel)
```

```python
# Apply erosion method
diaimg1=cv2.dilate(thresh,kernel,iterations=1)
diaimg2=cv2.dilate(thresh,kernel,iterations=2)
diaimg3=cv2.dilate(thresh,kernel,iterations=3)
```

```python
# Displaying the image
plt.subplot(231),plt.imshow(image, cmap='gray'),plt.title('Original Image'), plt.axis('off')
plt.subplot(232),plt.imshow(thresh, cmap='gray'),plt.title('Thresholded image'), plt.axis('off')
plt.subplot(234),plt.imshow(diaimg1, cmap='gray'),plt.title('Dilation i=1'), plt.axis('off')
plt.subplot(235),plt.imshow(diaimg2, cmap='gray'),plt.title('Dilation i=2'), plt.axis('off')
plt.subplot(236),plt.imshow(diaimg3, cmap='gray'),plt.title('Dilation i=3'), plt.axis('off')
plt.savefig("/content/lena color.png", format="png",dpi=1200)
plt.show()
```
