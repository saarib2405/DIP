```python
import cv2
import matplotlib.pyplot as plt
# load the input image
img = cv2.imread("/content/lena color.png")
# convert BGR to RGB to display using matplotlib
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#Apply Median Blur with Kernel sizes 3*3, 5*5, 7*7
medi3 = cv2.medianBlur(imgRGB,3)
medi5 = cv2.medianBlur(imgRGB,5)
medi7 = cv2.medianBlur(imgRGB,7)
plt.subplot(221),plt.imshow(imgRGB),plt.title('Original Image'), plt.axis('off')
plt.subplot(222),plt.imshow(medi3),plt.title('Median Blur (k=3*3)'), plt.axis('off')
plt.subplot(223),plt.imshow(medi5),plt.title('Median Blur (k=5*5)'), plt.axis('off')
plt.subplot(224),plt.imshow(medi7),plt.title('Median Blur (k=7*7)'), plt.axis('off')
plt.show()
```
